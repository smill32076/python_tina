### python如何讀取students.csv
你可以使用 Python 內建的 `csv` 模組來讀取 `students.csv`。以下是兩種常見的讀取方式：

---

## 方法一：使用 `csv.DictReader`（推薦）

這種方式會將每一列讀成字典，欄位名稱會自動對應到字典的 key，方便操作。

```python
import csv

with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # 將所有資料讀取成 list of dict
    students = [row for row in reader]

# 印出讀取結果
for student in students:
    print(student)
```


> 注意：讀取的數值都是字串，如果需要數字型態，可以額外轉換。

---

## 方法二：使用 `csv.reader`

這種方式會讀取成二維列表（list of list），第一列通常是欄位名稱。

```python
import csv

with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    
    # 讀取所有列
    rows = list(reader)

# 印出欄位名稱
header = rows[0]
print("欄位名稱:", header)

# 印出資料列
for row in rows[1:]:
    print(row)
```

---

## 補充：將字串轉成數字

如果你想把成績欄位轉成整數，可以這樣做：

```python
for student in students:
    student['國文'] = int(student['國文'])
    student['英文'] = int(student['英文'])
    student['數學'] = int(student['數學'])
```

## 總結

- **讀取 CSV 建議使用 `csv.DictReader`，方便以欄位名稱存取資料。**
- 記得指定 `encoding='utf-8'` 以支援中文。
- 若需要數字型態，讀取後可手動轉換。
